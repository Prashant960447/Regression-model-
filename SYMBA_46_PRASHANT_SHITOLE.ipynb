```python
def missing_data_threshold(data, Threshold = 75 ):
    missing_data = pd.DataFrame(data.isna().sum(), columns= ["missing_count"])
    missing_data["missing_percentage"] = missing_data["missing_count"]/len(data)
    columns_to_be_deleted = missing_data[missing_data["missing_percentage"]>= Threshold].index
    print("Deleted columns due to missing values> Threshold are-->",
          list(columns_to_be_deleted))
    data= data.drop(columns= columns_to_be_deleted)
    return data
```

```python
def missing_data_treatment(data):
   '''Treat missing values using median and mode'''
   for  col in data.columns:
      if data[col].dtype =='object':
        mode_value = data[col].mode()[0]
        print(col, "-->", data[col].dtype, "-->", mode_value)
        data[col] = data[col].fillna(mode_value)
      else:
        median_value = data[col].median()
        print(col, "-->", data[col].dtype, "-->", median_value)
        data[col] = data[col].fillna(median_value)
   return data
```

```python
data = missing_data_treatment(data)
```

```python
def feature_engineering(data):
    ''' Reduce variety of columns having more than 5% variety using bins'''
    for col in data.columns:
      if 100*data[col].nunique()/len(data)>5:
        if  data[col].dtype=='object':
            print('Skip --> manually reduce variety for', col)
        else:
            new_col_name=col + '_bucket'
            data[new_col_name] = pd.qcut(data[col], 4, labels=['b1','b2','b3', 'b4'])
            print('Auto reduced variety for', col)
    return data
data = feature_engineering(data)
```

```python
def label_encoding(data):
    '''Convert all alphabetical column to numeric column'''
    LN = LabelEncoder()
    for col in data.columns:
        if (data[col].dtype=='object') or (data[col].dtype == 'category'):
            print(col)
            print('before -->', data[col].unique())
            data[col] = LN.fit_transform(data[col])
            print('After -->', data[col].unique())
            print('-'*80)
    return data
```
